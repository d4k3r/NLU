Before running the cell below, upload the model card template (`COMP34812_modelcard_template.md`) provided to you using the Colab file browser (on the left-hand side).

```python
card_data = ModelCardData(
    language='en',
    license='apache-2.0',
    tags=['text-classification'],
    repo="https://github.com/d4k3r/NLU",
    ignore_metadata_errors=True)

card = ModelCard.from_template(
    card_data=card_data,
    template_path='COMP34812_modelcard_template.md',
    model_id='v25523st_m61733im_s70206jv-NLI',

    model_summary='''RWKV-7 "Goose" fine-tuned for Natural Language Inference. 
    Determines whether a hypothesis is entailed by a given premise.''',
    
    model_description='''This model fine-tunes RWKV-7 "Goose" (7.2B parameters) using LoRA
    for binary NLI classification. RWKV-7 is a recurrent architecture with linear time 
    complexity O(n) and constant space complexity O(1), achieving transformer-level 
    performance without attention mechanisms.''',
    
    developers='Serhii Tupikin, Ilya Maltsev, Julius Vander Arend',
    
    base_model_repo='https://huggingface.co/fla-hub/rwkv7-7.2B-g0a',
    base_model_paper='https://arxiv.org/abs/2503.14456',
    
    model_type='Supervised Fine-tuning with LoRA',
    model_architecture='RNN',
    language='English',
    base_model='fla-hub/rwkv7-7.2B-g0a',
    
    training_data='''
    COMP34812 NLU Coursework dataset:
    - 24,432 premise-hypothesis pairs for training
    - 6,736 pairs for validation''',
    
    hyperparameters='''
    - learning_rate: 1e-5
    - batch_size: 3
    - grad_accum_steps: 6
    - max_length: 320
    - num_epochs: 4
    - warmup_ratio: 0.1
    - lora_r: 32
    - lora_alpha: 64
    - lora_dropout: 0.05
    - lora_target_modules: [key, value, receptance, gate, output]
    - precision: BF16
    - optimizer: AdamW (weight_decay=0.01)''',
    
    speeds_sizes_times='''
    - overall training time: 2 hours
    - duration per training epoch: 30 minutes
    - model size: ~16GB (backbone + classifier)
    - trainable parameters: 83,886,080 (1.1518% of total)''',
    
    testing_data='''
    - Development set: 6,736 premise-hypothesis pairs''',
    
    testing_metrics='''
    - F1-score
    - Accuracy
    - Matthews Correlation Coefficient (MCC)''',
    
    results='''
    Dev set performance:
    - F1-score: 0.9419
    - Accuracy: 0.9399
    - MCC: 0.8796''',
    
    hardware_requirements='''
    - GPU: RTX5090 or equivalent GPU
    - RAM: 32GB recommended
    - Storage: 30GB for model and checkpoints''',
    
    software='''
    - Python 3.13+
    - PyTorch 2.11+
    - Triton 3.6+
    - Transformers 5.4+
    - PEFT 0.18+
    - Accelerate 1.13+
    - flash-linear-attention 0.4+ (for RWKV-7)
    - fla-core 0.4+
    - Pandas 3.0+
    - NumPy 2.4+
    - scikit-learn 1.8+
    - Matplotlib 3.10+''',

    bias_risks_limitations='''
    - Trained only on COMP34812 coursework data; may not generalize to other NLI datasets
    - English language only
    - Maximum sequence length: 320 tokens (longer inputs truncated)
    - Binary classification only (entailed vs. not entailed, no neutral class)
    - May inherit biases present in training data
    - Architecture Implementation Constraints: To ensure compatibility with the Hugging Face ecosystem and the PEFT library (LoRA), this model utilizes the `fla-hub` community implementation of RWKV-7. The original authors of RWKV note that the FLA port does not yet fully replicate the highly specific initialization rules and layer-specific learning rates of the native C++/CUDA reference code. Consequently, the model's current performance might be slightly below its theoretical maximum.
    - Training Dynamics: Despite the aforementioned implementation limitations, the 7.2B model demonstrated superior sample efficiency, successfully outperforming the 2.9B parameter variant in significantly fewer training epochs.''',
    
    additional_information='''
    - Uses last-token pooling for sequence representation
    - Classification head: LayerNorm -> Dropout(0.1) -> Linear
    - Model saved as: backbone/ (RWKV weights), classifier.pt, config.pt
    - Standalone evaluation cell available in notebook (Section 13)
    - Hyperparameters were determined by iterative manual testing
    
    References:
    - Peng et al. (2025). RWKV-7 "Goose" with Expressive Dynamic State Evolution. arXiv:2503.14456
    - Hu et al. (2021). LoRA: Low-Rank Adaptation of Large Language Models. arXiv:2106.09685'''
)

# Save model card
with open('model_card_rwkv7_nli.md', 'w') as model_card:
    model_card.write(card.content)
```
